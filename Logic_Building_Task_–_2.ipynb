```python
sentence = "python is easy and python is powerful"
sentence_list = list(sentence.split())
unique_words = []
for i in range(len(sentence_list)):
    if sentence_list[i] not in unique_words:
        unique_words.append(sentence_list[i])
print(len(unique_words))
print(unique_words)
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
if employees["Ravi"] > employees["Anita"] and employees["Ravi"] > employees["Kiran"]:
  print("Highest Salary: Ravi -", employees.get("Ravi"))
elif employees['Anita'] > employees["Ravi"] and employees["Anita"] > employees["Kiran"]:
  print("Highest Salary: Anita -", employees.get("Anita"))
else:
  print("Highest Salary: Kiran -", employees.get("Kiran"))
```

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for i in prices:
  if i > 1000:
    count += 1
print(count)
```
