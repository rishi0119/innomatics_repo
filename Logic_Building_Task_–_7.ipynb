```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(self.name, "is", self.status)

light = SmartLight(input())
action = input()
if action == "ON":
    light.turn_on()
else:
    light.turn_off()
light.display_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
emp_name = input()
emp_id = input()
emp_department = input()
emp = Employee(emp_name, emp_id, emp_department)
emp.display()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

name = input()
phone = input()
c = Contact(name, phone)
c.display()
```

```python
from IPython.core.prefilter import PrefilterChecker
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

product = input()
price = int(input())
p = Product(product, price)
p.display()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

name = input()
rating = float(input())
m = Movie(name, rating)
m.display()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

customer_name = input()
address = input()
d = Delivery(customer_name, address)
d.display()
```
