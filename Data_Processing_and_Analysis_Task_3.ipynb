```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
High_Score = max(employees.values())
for name, score in employees.items():
  if score == High_Score:
    print(f"Top Performers Eligible for Bonus: {name}, {score} (Score: {score})")
    break
```

```python
string = input()
string = string.lower()
string = list(string.split())
for i in set(string):
  c = string.count(i)
  if c > 1:
    print(f"{i}:{c}")
```

```python
sensor_readings = list(map(int, input().split(',')))
valid_readings = []
for i in range(len(sensor_readings)):
  if sensor_readings[i]%2==0:
    valid_readings.append((i, sensor_readings[i]))
print(f"Valid Sensor Readings (Hour, Value): {valid_readings}")
```

```python
emails = list(map(str, input().split(',')))
str1 = "@gmail.com"
str2 = "@yahoo.com"
str1_no = 0
str2_no = 0
for i in emails:
  if str1 in i:
    str1_no += 1
  elif str2 in i:
    str2_no += 1
  else:
    print("Enter valid mail ID")
str1_percentage = int((str1_no/len(emails))*100)
str2_percentage = int((str2_no/len(emails))*100)
print(f"gmail.com: {str1_percentage}% yahoo.com: {str2_percentage}%")
```

```python
sales = list(map(int, input().split(',')))
avg_sales = sum(sales)/len(sales)
r_sales = avg_sales+avg_sales*0.30 #here for required where the days of sales 30% above average should be displayes therefore the required sales is 2210 but in the questions expected output the day4 2200 is also given where it is not greter or equal to the required sales but nearest one
for i in range(len(sales)):
  if r_sales<sales[i]:
    print(f"Day{i+1}: {sales[i]}")
```
