```python
#1. User Login Check
username = "admin"
password = "1234"
user_name = input("Enter username: ")
pass_word = input("Enter password: ")
if user_name == username and pass_word == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
result = []
for i in marks:
  if i >= 50:
    result.append("pass")
  else:
    result.append("fail")
print(f"The total number of students passed are {result.count("pass")}")
print(f"The total number of students passed are {result.count("fail")}")
print(f"Therfore the result of the students are {result}")
```

```python
#3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for i in names:
  i = i.strip()
  i = i.lower()
  cleaned_names.append(i)
print(f"The cleaned data is {cleaned_names}")
```

```python
#4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
  print(f"The length of message {i} is {len(i)}")
  if len(i) > 10:
    print(f"The message {i} is greater than 10 characters.")
```
